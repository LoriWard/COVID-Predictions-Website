## COVIDP Data

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
csv_file = "covid_p.csv"
covidp = pd.read_csv(csv_file)
covidp.head()
```

```python
covidp.columns
```

```python
covidp = covidp.drop(columns=['Continent'])
covidp.head()
```

```python
#Data Analysis for US
us = covidp[covidp["Country"]=="US"]
us.head()
```

```python
us=us.fillna(0)
us.head()
```

```python
us.columns
```

```python
us = pd.DataFrame(us.groupby("Day").sum())
us.reset_index(inplace = True)
us.head()
```

```python
# Export the cleaned data as a csv file
us.to_csv("covp_us.csv")
 
```
